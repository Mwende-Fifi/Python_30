1. **Scatter Plot:** Visualize the relationship between `sepal_length` and `petal_length` in the `iris` dataset. 

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# datasets
iris = pd.read_csv('iris.csv', encoding_errors='ignore')
car_crashes = pd.read_csv('car_crashes.csv', encoding_errors='ignore')
flights = pd.read_csv('flights.csv', encoding_errors='ignore')
taxis = pd.read_csv('taxis.csv', encoding_errors='ignore')
tips = pd.read_csv('tips.csv', encoding_errors='ignore')
```

```python
iris.head()
```

```python
sns.scatterplot(data=iris, x='petal_length', y='sepal_length', hue='species')
```

2. **Heatmap:** Create a heatmap for the `flights` dataset showing passengers per month and year.  

```python
flights.head()
```

```python
# Pivot the data to create a matrix for the heatmap
flights_pivot = flights.pivot_table(index="year", columns="month", values="passengers")

# Ensure the months are in the correct order
month_order = [
    "January", "February", "March", "April", "May", "June",
    "July", "August", "September", "October", "November", "December"
]
flights_pivot = flights_pivot[month_order]  # Reorder columns manually

plt.figure(figsize=(10,6))
sns.heatmap(data=flights_pivot.corr(), annot=True)

# Add title and labels
plt.title("Heatmap of Passengers per Month and Year")
plt.xlabel("Month")
plt.ylabel("Year")

# Show the plot
plt.show()
```
